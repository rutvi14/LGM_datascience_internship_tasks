### Import libraries

```python
%config IPCompleter.greedy=True
import cv2
import matplotlib.pyplot as plt
```

```python
#read path adn set file name

img_loc = 'shivji.jpg'
```

```python
#read image
img = cv2.imread(img_loc)
```

```python
#show image
plt.figure(figsize=(10,10))

plt.imshow(img)
```

```python
#convert image to gray scale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(gray_image)
```

```python
#invert the image
inverted_gray_image = 255 - gray_image
```

```python
#cv2.imshow("New image",inverted_gray_image)
plt.figure(figsize=(8,8))
plt.imshow(inverted_gray_image)
```

Blur the image

```python
blurred_img = cv2.GaussianBlur(inverted_gray_image,(21,21),0)
```

```python
plt.figure(figsize=(8,8))
plt.imshow(blurred_img)
```

Invert the blurred image

```python
inverted_blurred_img= 255 - blurred_img
plt.figure(figsize=(8,8))
plt.imshow(inverted_blurred_img)
```

# Now create pencil sketch image

```python
pencil_sketch = cv2.divide(gray_image,inverted_blurred_img,scale=256.0)
plt.figure(figsize=(8,8))
plt.imshow(pencil_sketch)
```

```python
sketch = cv2.cvtColor(pencil_sketch, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(8,8))
plt.imshow(sketch)
```
